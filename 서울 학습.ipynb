```python
# study 데이터의 상권업종소분류명 기준으로 빈도수를 구해서
# study_count에 저장하고 출력하기

study_count = study["상권업종소분류명"].value_counts()
study_count
```

```python
# 그룹화 하기
# 시군구명과 상권업종소분류명을 기준으로 멀티 인덱스
# 상호명을 기준으로 갯수 세어주기
# 결과를 study_group에 저장

study_group = study.groupby(["시군구명", "상권업종소분류명"])["상호명"].count()
study_group
```

```python
study_group["강남구"]["태권도/무술학원"]
```

```python
# 그룹화 하기
# 상권업종소분류명, 시군구명을 기준으로 멀티 인덱스
# 상호명을 기준으로 갯수 세어주기
# 결과를 study_group에 저장

study_group2 = study.groupby(["상권업종소분류명", "시군구명"])["상호명"].count()
study_group2
```

```python
study_group2
```

```python
study_count_1000 = study_count[study_count >= 1000]
```

```python
max_counts = study_group2.groupby('상권업종소분류명').idxmax()
max_counts = max_counts.apply(lambda x: x[1])

result = max_counts[max_counts != '강남구']
print(result.index.tolist())
```

```python
# 멀티 인덱스로 되어 있는 데이터 인덱스 초기화 해주기
# 인덱스 초기화 된 데이터 t 변수에 저장해주기
# study_group 데이터 사용

t = study_group.reset_index()
t
```

```python
# t변수의 상호명 컬림을 상호수로 변경하기
t.rename(columns={'상호명': '상호수'}, inplace=True)
t
```
