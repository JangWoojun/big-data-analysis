```python
# csv 파일을 불러와서 df 변수에 저장하기
df = pd.read_csv('/content/drive/MyDrive/빅분/2024/소상공인시장진흥공단_상가(상권)정보_서울_202312 (1).xls',low_memory=False)
df
```

```python
# ① 컬럼을 변환할 수 있다.
# ② 컬럼 구조를 변환할 수 있다.
# ③ 데이터를 그룹화할 수 있다.
# ④ 피봇 테이블을 활용할 수 있다.
# ⑤ 비율과 수치를 활용하여 데이터를 검증할 수 있다.
# ⑥ 시각화를 활용하여 데이터를 검증할 수 있다.
```

```python
# ① 컬럼을 변환할 수 있다.
df.columns
df.rename(columns={"상호명" : "상호 이름"}, inplace=True)
```

```python
df.columns
```

```python
df["시도코드"].dtype
```

```python
# ② 컬럼 구조를 변환할 수 있다.
df["시도코드"] = df["시도코드"].astype(str)
df["시도코드"].dtype
```

```python
# ③ 데이터를 그룹화할 수 있다.
df.groupby(['시군구명','상권업종소분류명'])['상호 이름'].count()
```

```python
# ④ 피봇 테이블을 활용할 수 있다.
df.pivot_table(index=["시군구명", "상권업종소분류명"], values="상호 이름", aggfunc="sum")
```

```python
# ⑤ 비율과 수치를 활용하여 데이터를 검증할 수 있다.
df[df["시군구명"] == "은평구"]["법정동명"].value_counts(normalize=True)
```

```python
# ⑥ 시각화를 활용하여 데이터를 검증할 수 있다.
plt.figure(figsize = (15, 5))
sns.barplot(data = df, x = '시군구명', y = '상호 이름', errorbar = None)
```

```python
sns.catplot(data = df, x = '상권업종소분류명', y = '상호 이름', errorbar = None, aspect = 3, kind = 'bar')
```
