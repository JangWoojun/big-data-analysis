```python
df_food = df[df["상권업종대분류명"] == "음식"]
```

```python
# 강남구에 있는 상가 데이터만 가져오기

df[df["시군구명"] == "강남구"]
```

```python
# 상권업종대분류명이 음식이면서 강남구에 있는 상점 데이터 가져오기

df[(df["시군구명"] == "강남구") & (df["상권업종대분류명"] == "음식")]
```

```python
df[(df["시군구명"] == "강남구") & (df["상권업종대분류명"] == "음식")]["상권업종중분류명"].value_counts()
```

```python
# Loc를 이용해서 행, 열 데이터를 가져오면 처리속도가 빠름 -> 데이터 프레임에 접근하는 횟수 차이
# LOC [행, 열]
df.loc[(df["시군구명"] == "강남구") & (df["상권업종대분류명"] == "음식"), "상권업종중분류명"].value_counts()
```

```python
# 1차 시군구명으로 분류, 2차 상권업종중분류명으로 분류해서 갯수 세기
food_gu = df_food.groupby(["시군구명", "상권업종중분류명"])["상호명"].count()
food_gu
```

```python
# 피봇 테이블로 변경하기
# columns = "컬럼명"
# values = "연산에 사용할 값"
# aggfunc = "연산"
# 연산에 sum을 넣어주면 문자 데이터기 때문에 모든 문자들이 연결되서 출력됨

df_food.pivot_table(index=["시군구명", "상권업종중분류명"], values=["상호명"], aggfunc = "count")
```

```python
df_food.pivot_table(index=["시군구명", "상권업종중분류명"], values=["상호명"], aggfunc = "count").sort_values(by=["시군구명", "상호명"], ascending = [True, False])
```

```python
food_gu.plot(figsize = (15,5))
```

```python
# unstack 하면 멀티 인덱스 (시군구명, 상권 업종중분류명)형태로 되어있던 형태에서 
# 마지막 인덱스에 있던 상건업종중분류명이 컬럼으로 가게 됨

food_gu.unstack()
```
