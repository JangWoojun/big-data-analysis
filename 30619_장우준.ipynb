```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# csv 파일을 불러와서 store 변수에 저장하기
# 파일경로 './소상공인시장진흥공단_상가(상권)정보_서울_202312.csv'

store = pd.read_csv('./소상공인시장진흥공단_상가(상권)정보_서울_202312.csv')
```

```python
# 결측치 있는 값 정렬해서 상위 7개 인덱스 저장

not_use = n.sort_values(ascending = False).head(7).index
not_use
```

```python
# not_use를 사용해 결측치 많은 열 삭제

print(store.shape)
store = store.drop(not_use, axis = 1)
print(store.shape)
```

```python
# store 정보 출력

store.info()
```

```python
# store columns 변수로 저장

cols = store.columns
cols
```

```python
# store에 column 중에서 코드 또는 번호를 포함한 column을 변수에 저장 후 출력

not_cols = cols[cols.str.contains('코드|번호')]
not_cols
```

```python
# 아까 저장했던 변수를 이용해 해당 열 삭제

print(store.shape)
store.drop(not_cols, axis = 1, inplace = True)
print(store.shape)
```

```python
# store 정보 출력

store.info()
```

```python
# 0부터 6까지 행과 마지막부터 뒤에서 7번째 있는 열 출력
store.iloc[:7, -7:]
```

```python
# 위의 결과를 loc를 이용하여 출력하기
store.loc[:6, ["지번본번지",	"지번주소",	"도로명",	"건물본번지",	"도로명주소",	"경도",	"위도"]]
```

```python
# store에서 상권업종대분류명이 음식인 애들을 복사해서 변수에 저장

store_food = store[store['상권업종대분류명'] == '음식'].copy()
store_food
```

```python
# 강남구에 있는 음식점들을 상권업종중분류명을 기준으로 그룹화해서 갯수 출력하기

store_food[store_food['시군구명'] == '강남구'].groupby('상권업종중분류명').count()
```

```python
# 상권업종대분류명이 음식인 애들에서 시군구명과 상권업종중분류명을 groupby해서 상호명 갯수를 출력

food_gu = store_food.groupby(['시군구명', '상권업종중분류명'])['상호명'].count()
food_gu
```

```python
# pivot_table로 시군구명 별 상권업종중분류명 갯수를 상호명 기준으로 계산

store_food.pivot_table(index = ['시군구명', '상권업종중분류명'],
                    values = ['상호명'],
                    aggfunc = 'count')
```

```python
# 상권업종대분류명이 음식인 애들에서 시군구명과 상권업종중분류명을 groupby해서 상호명 갯수를 시각화

food_gu.plot(figsize = (15, 7))
```

```python
# 멀티인덱스에서 젤 마지막에 넣은 것을 뺀다

food_gu.unstack()
```

```python
# 멀티인덱스에서 젤 마지막에 넣은 것을 빼면서 값을 시각화

food_gu.unstack().plot.bar(figsize = (15, 7))
```

```python
# 멀티인덱스에서 젤 마지막에 넣은 것을 빼고 강남구 값을 출력

food_gu.unstack().loc['강남구']
```

```python
# 멀티인덱스에서 젤 마지막에 넣은 것을 빼고 강남구 값을 시각화

food_gu.unstack().loc['강남구'].plot()
```

```python
# food 변수에 상권업종대분류명이 음식인 애들에서 시군구명과 상권업종중분류명을 groupby해서 상호명 갯수 값을 reset_index 해서 저장

food = food_gu.reset_index()
food
```

```python
# food에 column 중 상호명을 상호수로 변경 후 출력

food = food.rename(columns = {'상호명':'상호수'})
food
```

```python
# 도화지 사이즈를 설정 후 막대 그래프로 시군구명별 상호수를 시각화
# 데이터는 food로

plt.figure(figsize = (15, 4))
sns.barplot(data = food, x = '시군구명', y = '상호수')
```

```python
# 도화지 사이즈를 설정 후 막대 그래프로 시군구명별 상호수, 범례를 상권업종중분류명로 시각화
# 데이터는 food로

plt.figure(figsize = (20, 5))
sns.barplot(data = food, x = '시군구명', y = '상호수',
            hue = '상권업종중분류명')
```

```python
# 카테고리형 그래프로 시군구명별 상호수로 그래프를 2배로 시각화
# 데이터는 food로

sns.catplot(data = food, x = '시군구명', y = '상호수', aspect = 2)
```

```python
# 카테고리형 그래프로 시군구명별 상호수로 그래프를 3배로 막대 그래프로 시각화
# 데이터는 food로

sns.catplot(data = food, x = '시군구명', y = '상호수',
            kind = 'bar', aspect = 3)
```

```python
# 카테고리형 그래프로 시군구명별 상호수로 그래프를 3배로 막대 그래프로 범례는 시군구명, x, y 텍스트를 모두 보여주게 설정해주고 색은 Set2로 열은 상권업종중분류명로 2줄로 시각화
# 데이터는 food로

sns.catplot(data = food, x = '시군구명', y = '상호수', kind = 'bar',
            aspect = 3, col = '상권업종중분류명', col_wrap = 2, hue = '시군구명',
            sharex = False, sharey = False, palette = 'Set2')
```

```python
# food 데이터를 정렬하여 food_sort에 저장
# 정렬기준 시군구명(내림차순), 상호수(오름차순)
food_sort = food.sort_values(by = ["시군구명", "상호수"], ascending=[False, True])
# food_sort 출력
food_sort
```

```python
# 카테고리형 그래프로 상권업종중분류명별 상호수로 그래프를 3배로 막대 그래프로 범례는 상권업종중분류명, x, y 텍스트를 모두 보여주게 설정해주고 색은 Set2로 열은 시군구명로 2줄로 시각화
# 데이터는 food로

sns.catplot(data = food_sort, x = '상권업종중분류명', y = '상호수', kind = 'bar',
            aspect = 2, col = '시군구명', col_wrap = 2, hue = '상권업종중분류명',
            sharex = False, sharey = False, palette = 'Set2')
```
