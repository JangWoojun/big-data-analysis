```python
import re
```

```python
name = "CarizarMega Charizard X"
```

```python
name_split = name.split()
name_split
```

```python
tokens = re.findall("[A-Z][a-z]*", name_split[0])
tokens
```

```python
tokens = []

for i in name_split:
    tokens.extend(re.findall("[A-Z][a-z]*", i))

tokens 
```

```python
def tokenize(name): # tokenize 함수
    name_split = name.split()
    tokens = []
    for i in name_split:
        tokens.extend(re.findall("[A-Z][a-z]*", i))
    return tokens
```

```python
tokenize(name) # tokenize 함수 테스트
```

```python
l_pokemon["Name"].apply(tokenize) # l_pokemon
```
